### Limpieza de los datasets

```python
# Limpieza de CONSUMO

# Eliminar columnas no necesarias
columnas_relevantes = ['Fecha', 'Diesel bajo azufre', 'Diesel ultra bajo azufre', 'Gas licuado de petróleo',
                       'Gasolina regular', 'Gasolina superior', 'Diesel alto azufre']
consumo_df = consumo_df[columnas_relevantes]

# Renombrar columnas
consumo_df.rename(columns={
    'Fecha': 'Fecha',
    'Diesel bajo azufre': 'DieselBajo',
    'Diesel ultra bajo azufre': 'DieselUltraBajo',
    'Gas licuado de petróleo': 'GasLicuado',
    'Gasolina regular': 'Regular',
    'Gasolina superior': 'Superior',
    'Diesel alto azufre': 'DieselAlto'
}, inplace=True)

# Eliminar comas de las columnas y convertir a float
consumo_df['DieselBajo'] = consumo_df['DieselBajo'].replace(',', '', regex=True).astype(float)
consumo_df['DieselUltraBajo'] = consumo_df['DieselUltraBajo'].replace(',', '', regex=True).astype(float)
consumo_df['DieselAlto'] = consumo_df['DieselAlto'].replace(',', '', regex=True).astype(float)


# Union de las variantes de Diesel
consumo_df['Diesel'] = consumo_df['DieselBajo'] + consumo_df['DieselUltraBajo'] + consumo_df['DieselAlto']

# Eliminar las columnas de variantes de Diesel
consumo_df.drop(columns=['DieselBajo', 'DieselUltraBajo', 'DieselAlto'], inplace=True)

# Convertir las columnas a tipo float
consumo_df['GasLicuado'] = consumo_df['GasLicuado'].replace(',', '', regex=True).astype(float)
consumo_df['Regular'] = consumo_df['Regular'].replace(',', '', regex=True).astype(float)
consumo_df['Superior'] = consumo_df['Superior'].replace(',', '', regex=True).astype(float)

# Estandarizar fecha con el formato MM/YYYY
consumo_df['Fecha'] = pd.to_datetime(consumo_df['Fecha']).dt.strftime('%m/%Y')

display(consumo_df)
print("\n")
print("TIPOS DE DATOS CONSUMO")
print(consumo_df.dtypes)
print("\n")
print("DATOS NULOS")
print(consumo_df.isnull().sum())
```

#### Observaciones
- No se conviertieron a int si no que a float para mantener la exactitud de los datos.
- Al colocar la `Fecha` como index, presenta un error y crea una fila extra. Se resolvera mas adelante si es que esta columna default index presenta algun problema.

```python
# Guardar el archivo limpio
consumo_df.to_csv('data/cleanData/CONSUMO.csv', index=False)
```

### Análisis de Importaciones

```python
# Convertir la columna 'Fecha' a formato datetime (con formato '%m/%Y')
importacion_df['Fecha'] = pd.to_datetime(importacion_df['Fecha'], format='%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if importacion_df['Fecha'].isnull().any():
    print("Hay valores nulos en la columna 'Fecha' después de la conversión.")

# Extraer el mes y el año
importacion_df['Month'] = importacion_df['Fecha'].dt.to_period('M')
importacion_df['Year'] = importacion_df['Fecha'].dt.year

# Total de importaciones por mes
importacion_df['Month_Name'] = importacion_df['Fecha'].dt.strftime('%B')

# Agrupar por mes y calcular el promedio de importaciones
importaciones_mensuales = importacion_df.groupby('Month_Name').mean().sort_index()

# Ordenar los meses en orden cronológico
meses_ordenados = ['January', 'February', 'March', 'April', 'May', 'June', 
                   'July', 'August', 'September', 'October', 'November', 'December']
importaciones_mensuales = importaciones_mensuales.reindex(meses_ordenados)

# Graficar las importaciones promedio por mes para cada tipo de combustible
importaciones_mensuales.plot(kind='bar', figsize=(14, 7))
plt.title('Promedio de Importaciones por Mes y Tipo de Combustible')
plt.xlabel('Mes')
plt.ylabel('Importación Promedio')
plt.xticks(rotation=45)
plt.legend(title='Tipo de Combustible')
plt.show()


# Total de importaciones por año
importaciones_anuales = importacion_df.groupby('Year').sum()
importaciones_anuales.plot(kind='bar', figsize=(12, 6))
plt.title('Total de Importaciones por Año')
plt.ylabel('Importación Total')
plt.show()
```

### Comportamiento Durante la Pandemia de la Importacion

```python
# Definir el inicio y fin de la pandemia
pandemia_inicio = pd.Timestamp('2020-03-13')
pandemia_fin = pd.Timestamp('2022-06-01')

# Dividir los datos en pre-pandemia, durante pandemia, y post-pandemia
pre_pandemia = importacion_df[importacion_df['Fecha'] < pandemia_inicio]
durante_pandemia = importacion_df[(importacion_df['Fecha'] >= pandemia_inicio) & (importacion_df['Fecha'] <= pandemia_fin)]
post_pandemia = importacion_df[importacion_df['Fecha'] > pandemia_fin]

# Calcular el promedio de importaciones para cada fase
pre_pandemia_mean = pre_pandemia[['GasLicuado', 'Regular', 'Superior', 'Diesel']].mean()
durante_pandemia_mean = durante_pandemia[['GasLicuado', 'Regular', 'Superior', 'Diesel']].mean()
post_pandemia_mean = post_pandemia[['GasLicuado', 'Regular', 'Superior', 'Diesel']].mean()

# Crear un DataFrame para la comparación
comparison = pd.DataFrame({
    'Pre-Pandemia': pre_pandemia_mean,
    'Durante Pandemia': durante_pandemia_mean,
    'Post-Pandemia': post_pandemia_mean
})

# Graficar la comparación de importaciones en las diferentes fases de la pandemia
comparison.plot(kind='bar', figsize=(12, 6))
plt.title('Comparación de Importaciones Pre, Durante y Post Pandemia')
plt.ylabel('Promedio de Importación')
plt.xticks(rotation=45)
plt.show()
```

Aqui es facil ver que el consumo de todos los tipos de combustible son altos, y esa es la correlacion alta que se ve uno con otros.

- Importacion:
Vemos que se casi siempre se importan las mismas cantidades de regular y super.

Se ve que los precios del Diesel y el Gas Licuado no varian un poco, sin embargo no demasiado, como ya lo habiamos observado antes. Mientras que los precios de los combustibles superior y regular son casi completamente exactos.

```python
# Combinar los DataFrames basados en la columna 'Fecha'
combined_df = pd.merge(consumo_df, importacion_df, on='Fecha', suffixes=('_consumo', '_importacion'))

# Calcular la correlación entre las variables de consumo y de importación
corr_combined = combined_df[['GasLicuado_consumo', 'Regular_consumo', 'Superior_consumo', 'Diesel_consumo',
                             'GasLicuado_importacion', 'Regular_importacion', 'Superior_importacion', 'Diesel_importacion']].corr()

# Visualizar la correlación entre consumo e importación
plt.figure(figsize=(12, 10))
sns.heatmap(corr_combined, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlación entre Consumo e Importación')
plt.show()
```

#### Observaciones
1. Consumo e Importación del Combustible Regular:
La alta correlación entre el consumo y la importación del combustible regular sugiere que este tipo de combustible tiene una coherencia significativa entre la oferta y la demanda en el mercado. Esto indica que las cantidades importadas de combustible regular están bien alineadas con su consumo real, lo cual puede reflejar una planificación y gestión más precisa de este recurso.

2. Gas Licuado y Combustible Regular:
El gas licuado muestra una tendencia de consumo similar a la del combustible regular. Esta similitud en los patrones de consumo podría estar relacionada con los usos y aplicaciones comunes de ambos tipos de combustibles, sugiriendo que los consumidores pueden utilizar ambos combustibles de manera intercambiable en algunos contextos.

3. Correlación entre Consumo del Combustible Regular y Superior:
Existe una alta correlación entre el consumo de combustible regular y superior. Esta alta correlación puede atribuirse a que ambos combustibles tienen un uso extensivo y pueden ser empleados en una variedad de aplicaciones donde el consumo tiende a ser elevado. La similitud en el consumo de estos combustibles podria reflejar una demanda conjunta en ciertos segmentos del mercado.

4. Consumo de Diesel y Combustible Regular:
El consumo de diesel y combustible regular muestra una correlación moderada, lo que indica que, aunque ambos tipos de combustibles tienen patrones de consumo relativamente similares, las diferencias en su uso específico (como en vehículos pesados para el diesel y vehículos ligeros para el combustible regular) pueden influir en esta relación.
