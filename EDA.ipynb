### Limpieza de los datasets

```python
# Limpieza de CONSUMO

# Eliminar columnas no necesarias
columnas_relevantes = ['Fecha', 'Diesel bajo azufre', 'Diesel ultra bajo azufre', 'Gas licuado de petróleo',
                       'Gasolina regular', 'Gasolina superior', 'Diesel alto azufre']
consumo_df = consumo_df[columnas_relevantes]

# Renombrar columnas
consumo_df.rename(columns={
    'Fecha': 'Fecha',
    'Diesel bajo azufre': 'DieselBajo',
    'Diesel ultra bajo azufre': 'DieselUltraBajo',
    'Gas licuado de petróleo': 'GasLicuado',
    'Gasolina regular': 'Regular',
    'Gasolina superior': 'Superior',
    'Diesel alto azufre': 'DieselAlto'
}, inplace=True)

# Eliminar comas de las columnas y convertir a float
consumo_df['DieselBajo'] = consumo_df['DieselBajo'].replace(',', '', regex=True).astype(float)
consumo_df['DieselUltraBajo'] = consumo_df['DieselUltraBajo'].replace(',', '', regex=True).astype(float)
consumo_df['DieselAlto'] = consumo_df['DieselAlto'].replace(',', '', regex=True).astype(float)


# Union de las variantes de Diesel
consumo_df['Diesel'] = consumo_df['DieselBajo'] + consumo_df['DieselUltraBajo'] + consumo_df['DieselAlto']

# Eliminar las columnas de variantes de Diesel
consumo_df.drop(columns=['DieselBajo', 'DieselUltraBajo', 'DieselAlto'], inplace=True)

# Convertir las columnas a tipo float
consumo_df['GasLicuado'] = consumo_df['GasLicuado'].replace(',', '', regex=True).astype(float)
consumo_df['Regular'] = consumo_df['Regular'].replace(',', '', regex=True).astype(float)
consumo_df['Superior'] = consumo_df['Superior'].replace(',', '', regex=True).astype(float)

display(consumo_df)
print("\n")
print("TIPOS DE DATOS CONSUMO")
print(consumo_df.dtypes)
print("\n")
print("DATOS NULOS")
print(consumo_df.isnull().sum())
```

#### Observaciones
- No se conviertieron a int si no que a float para mantener la exactitud de los datos.
- Al colocar la `Fecha` como index, presenta un error y crea una fila extra. Se resolvera mas adelante si es que esta columna default index presenta algun problema.

```python
# Guardar el archivo limpio
consumo_df.to_csv('data/cleanData/CONSUMO.csv', index=False)
```
